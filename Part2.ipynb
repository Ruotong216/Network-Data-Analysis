```python
import osmnx as ox
import networkx as nx
import math
import geopandas as gpd
import matplotlib.pyplot as plt

latitude = 53.7975
longitude = -1.5429

G = ox.graph_from_point((latitude, longitude), dist=500, network_type='drive', simplify=True)
G_proj = ox.project_graph(G)

G_sub = max((G_proj.subgraph(c).copy() for c in nx.strongly_connected_components(G_proj)), key=len)
undirected_G = G_sub.to_undirected()

is_planar, _ = nx.check_planarity(undirected_G)

try:
    spatial_diameter = nx.diameter(undirected_G)
except Exception:
    spatial_diameter = "Not computable"

stats = ox.basic_stats(G_sub)

avg_street_length = stats['edge_length_avg']
intersection_count = stats['intersection_count']
avg_circuity = stats['circuity_avg']

area_km2 = math.pi * (0.5 ** 2)
node_density = G_sub.number_of_nodes() / area_km2
edge_density = stats['edge_length_total'] / 1000 / area_km2

print("Road Network Analysis (500m radius):")
print(f"- Spatial Diameter: {spatial_diameter}")
print(f"- Avg Street Length: {avg_street_length:.2f} m")
print(f"- Intersection Count: {intersection_count}")
print(f"- Avg Circuity: {avg_circuity:.3f}")
print(f"- Node Density: {node_density:.2f} nodes/km²")
print(f"- Edge Density: {edge_density:.2f} km/km²")
print(f"- Is Network Planar? {'Yes' if is_planar else 'No'}")

ox.plot_graph(G, bgcolor='white', node_size=10, edge_linewidth=0.5)
```

```python
import geopandas as gpd
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist

def ripley_k(points, support, area):
    """
    points: numpy  (n,2)  meters
    support: numpy
    area: m2
    """
    n = len(points)
    if n < 2:
        return np.zeros_like(support)

    dists = pdist(points)
    k_values = []
    for r in support:
        count = np.sum(dists <= r)
        # K(r) = (area / (n*(n-1))) * count
        K_r = area * count / (n * (n - 1))
        k_values.append(K_r)
    return np.array(k_values)


accidents = gpd.read_file("data/accidents_2015_2019_merged.gpkg", engine="fiona")

accidents = accidents.to_crs(epsg=27700)

points = np.array([(geom.x, geom.y) for geom in accidents.geometry])

r_max = 500  # meters
area_study = math.pi * (r_max ** 2) 

support = np.arange(0, r_max + 1, 10)

K_obs = ripley_k(points, support, area_study)

K_theo = math.pi * (support ** 2)

plt.figure(figsize=(8, 6))
plt.plot(support, K_obs, label="Observed K", color="blue")
plt.plot(support, K_theo, 'r--', label="Theoretical CSR")
plt.xlabel("Distance (m)")
plt.ylabel("K(r)")
plt.title("Manual Calculation of Ripley's K Function")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import osmnx as ox
import geopandas as gpd
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from shapely.geometry import LineString, Point
from libpysal.weights import W
from esda.moran import Moran
import networkx as nx

latitude = 53.7975
longitude = -1.5429
G = ox.graph_from_point((latitude, longitude), dist=500, network_type='drive', simplify=True)
G_proj = ox.project_graph(G)

edges = ox.graph_to_gdfs(G_proj, nodes=False, edges=True, node_geometry=False, fill_edge_geometry=True)
edges = edges.reset_index(drop=True)
edges["eid"] = edges.index

print("Edges columns:", edges.columns)

edges["buffer_geom"] = edges["geometry"].buffer(5)


accidents_gdf = accidents
accidents_gdf = accidents_gdf.to_crs(edges.crs)

edges_buffer = edges.copy().set_geometry("buffer_geom")
joined = gpd.sjoin(accidents_gdf, edges_buffer, how="left", predicate="intersects")

acc_count_series = joined.groupby("eid").size()
edges["accidents"] = edges["eid"].map(acc_count_series).fillna(0)

print("Accident counts summary:")
print(edges["accidents"].describe())
print(edges["accidents"].value_counts())

def get_endpoints(line):
    coords = list(line.coords)
    return {(coords[0][0], coords[0][1]), (coords[-1][0], coords[-1][1])}

edges["endpoints"] = edges["geometry"].apply(get_endpoints)

edge_neighbors = {eid: set() for eid in edges["eid"]}
for i in edges.index:
    for j in range(i+1, len(edges)):
        if len(edges.at[i, "endpoints"] & edges.at[j, "endpoints"]) > 0:
            edge_neighbors[edges.at[i, "eid"]].add(edges.at[j, "eid"])
            edge_neighbors[edges.at[j, "eid"]].add(edges.at[i, "eid"])

w = W(edge_neighbors)

y = edges["accidents"].values
print("y values:", y)

if np.var(y) == 0:
    print("Variance of accident counts is zero. Moran's I cannot be computed.")
else:
    mi = Moran(y, w)
    print(f"Moran's I: {mi.I:.4f}")
    print(f"p-value: {mi.p_sim:.4f}")
    

    if hasattr(mi, "sim") and len(mi.sim) > 0:
        plt.hist(mi.sim, bins=30, alpha=0.7, label="Simulated I")
        plt.axvline(mi.I, color='red', linestyle='--', label=f"Observed I = {mi.I:.4f}")
        plt.title("Moran's I Simulation Distribution")
        plt.xlabel("Moran's I")
        plt.ylabel("Frequency")
        plt.legend()
        plt.show()
```

```python
import osmnx as ox
import geopandas as gpd
import networkx as nx
from shapely.geometry import LineString, Point
import matplotlib.pyplot as plt

gpd.options.io_engine = "fiona"

place_name = "Leeds, UK"
G = ox.graph_from_place(place_name, network_type="drive")
G_proj = ox.project_graph(G, to_crs="EPSG:27700")

edges = ox.graph_to_gdfs(G_proj, nodes=False, edges=True)
edges = edges.reset_index(drop=True)
edges["eid"] = edges.index

voronoi = gpd.read_file("data/voronoi_polygons_v22.gpkg")
voronoi = voronoi.to_crs(edges.crs)

for idx, row in voronoi.iterrows():
    cell_id = row["id"] if "id" in row else idx
    polygon = row.geometry
    print(f"\n▶️ Processing cell {cell_id}...")

   
    edges_in_cell = edges[edges.geometry.intersects(polygon)].copy()

    if edges_in_cell.empty:
        print(f"  ⚠️ No edges found in cell {cell_id}")
        continue

   
    G_cell = nx.Graph()

    for _, edge in edges_in_cell.iterrows():
        geom = edge.geometry
        if not geom or geom.geom_type != "LineString":
            continue
        coords = list(geom.coords)
        for i in range(len(coords) - 1):
            u = coords[i]
            v = coords[i + 1]
            length = LineString([u, v]).length
            G_cell.add_edge(u, v, length=length, geometry=LineString([u, v]))

    if len(G_cell) == 0:
        print(f"  ⚠️ Could not build a graph in cell {cell_id}")
        continue

    
    cycles = list(nx.cycle_basis(G_cell))

    if not cycles:
        print(f"  ❌ No cycles found in cell {cell_id}")
        continue

    found_in_cell = False 

    for c_idx, cycle in enumerate(cycles):
        total_len = 0
        edges_geom = []

        for i in range(len(cycle)):
            u = cycle[i]
            v = cycle[(i + 1) % len(cycle)]
            if G_cell.has_edge(u, v):
                total_len += G_cell[u][v]["length"]
                edges_geom.append(G_cell[u][v]["geometry"])
            elif G_cell.has_edge(v, u):  # fallback
                total_len += G_cell[v][u]["length"]
                edges_geom.append(G_cell[v][u]["geometry"])

        
        if 41000 <= total_len <= 43000:
            print(f"    ✅ Cycle {c_idx + 1}: {total_len:.2f} meters (≈{total_len/1000:.2f} km)")
            gpd.GeoSeries(edges_geom).plot()
            plt.title(f"Cell {cell_id} - Cycle {c_idx+1} ({total_len/1000:.2f} km)")
            plt.axis("equal")
            plt.show()
            found_in_cell = True

    if not found_in_cell:
        print(f"    ❌ No ~42km cycle found in cell {cell_id}")

```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

agents = ["Council", "Volunteers", "Runners", "Medical Team", "Timing System"]
entities = [
    "Route Map", "Bib List", "Bib", "Checkpoint Data 1", "Checkpoint Data 2",
    "Result File", "Photos", "Water Bottles", "Medical Report", "Final Certificate"
]
activities = [
    "Plan Route", "Distribute Bibs", "Run Marathon", "Process Results", 
    "Distribute Water", "Handle Medical", "Generate Certificates"
]


G.add_nodes_from(agents, type="Agent")
G.add_nodes_from(entities, type="Entity")
G.add_nodes_from(activities, type="Activity")

edges = [
    ("Council", "Plan Route"),
    ("Plan Route", "Route Map"),
    ("Volunteers", "Distribute Bibs"),
    ("Bib List", "Distribute Bibs"),
    ("Route Map", "Distribute Bibs"),
    ("Distribute Bibs", "Bib"),
    ("Distribute Bibs", "Volunteers"),
    ("Bib", "Run Marathon"),
    ("Runners", "Run Marathon"),
    ("Run Marathon", "Checkpoint Data 1"),
    ("Run Marathon", "Checkpoint Data 2"),
    ("Run Marathon", "Photos"),
    ("Checkpoint Data 1", "Process Results"),
    ("Checkpoint Data 2", "Process Results"),
    ("Process Results", "Result File"),
    ("Result File", "Generate Certificates"),
    ("Generate Certificates", "Final Certificate"),
    ("Volunteers", "Distribute Water"),
    ("Distribute Water", "Water Bottles"),
    ("Run Marathon", "Handle Medical"),
    ("Medical Team", "Handle Medical"),
    ("Handle Medical", "Medical Report"),
    ("Timing System", "Process Results"),
]

G.add_edges_from(edges)


#pos = nx.spring_layout(G, seed=42)
pos = nx.spring_layout(G, seed=42, k=1.8, iterations=100)

colors = []
for node in G.nodes(data=True):
    if node[1]["type"] == "Agent":
        colors.append("orange")
    elif node[1]["type"] == "Entity":
        colors.append("lightgreen")
    else:
        colors.append("skyblue")

plt.figure(figsize=(14, 10))
nx.draw(G, pos, with_labels=True, node_color=colors, node_size=1300, edge_color="gray", arrows=True)
plt.title("PROV Network for Marathon Event (≥ 20 Nodes)", fontsize=14)
plt.axis("off")
plt.show()
```

```python
pr = nx.pagerank(G)

top_pr = sorted(pr.items(), key=lambda x: x[1], reverse=True)[:10]
print("Top 10 PageRank nodes:")
for node, score in top_pr:
    print(f"{node}: {score:.4f}")
```

```python
pr = nx.pagerank(G)

# different type
color_map = {
    "Activity": "#84c1ff",
    "Entity": "#aaffaa",
    "Agent": "#ffb347"
}

node_colors = []
node_sizes = []
for node in G.nodes(data=True):
    n_type = node[1].get("type", "entity") 
    node_colors.append(color_map.get(n_type, "#cccccc"))
    node_sizes.append(pr[node[0]] * 5000) 

plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42, k=1.8, iterations=100)

# nodes and edges
nx.draw_networkx_edges(G, pos, alpha=0.3)
nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=node_sizes)
nx.draw_networkx_labels(G, pos, font_size=8)

plt.title("Provenance Network with PageRank (Node Size = PR Score)")
plt.axis("off")
plt.tight_layout()
plt.show()
```
